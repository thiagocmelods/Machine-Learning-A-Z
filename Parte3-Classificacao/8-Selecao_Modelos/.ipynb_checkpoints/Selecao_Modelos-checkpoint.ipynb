# Importação de Bibliotecas

```python
# Importação de Bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Modelo de Regressão Logística

```python
# Carregando o dataset

dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Dividindo o dataset em dados de treino e dados de teste

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 0)
```

```python
# Dimensionsamento de recursos

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Construção do Modelo de Regressão Logística

from sklearn.linear_model import LogisticRegression
clf_logistica = LogisticRegression(random_state = 0)
```

```python
# Treinamento do Modelo de Regressão Logística

clf_logistica.fit(X_train, y_train)
```

```python
# Previsão do Modelo de Regressão Logística nos dados de teste

previsoes_logistica = clf_logistica.predict(X_test)
```

```python
# Matriz de Confusão do Modelo de Regressão Logística

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, previsoes_logistica)
```

```python
# Acurácia do Modelo de Regressão Logística

from sklearn.metrics import accuracy_score
accuracy_score(y_test, previsoes_logistica)
```

```python
# Visualização dos resultados do Modelo de Regressão Logística nos dados de teste

plt.rcParams['figure.figsize'] = (20,12)
plt.rcParams['font.size'] = 16
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['axes.labelsize'] = 20

tamanho = len(y_test)
plt.scatter([i for i in range(0,tamanho) if previsoes_logistica[i] == y_test[i]], previsoes_logistica[previsoes_logistica==y_test], s=50, color='green', label='Previsões Corretas')
plt.scatter([i for i in range(0,tamanho) if previsoes_logistica[i] != y_test[i]], previsoes_logistica[previsoes_logistica!=y_test], s=50, color='red', label='Previsões Incorretas')
plt.title('Resultado do Modelo com Regressão Logística ({} primeiras instâncias)'.format(tamanho))
plt.xlabel('Instãncias')
plt.xticks(range(0, tamanho))
plt.xticks(rotation=90)
plt.yticks(range(0, 2))
plt.ylabel('Tipo: 2 - Benígno / 4 - Malígno')
plt.legend()
plt.show()
```

# Modelo KNN

```python
# Carregando o dataset

dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Visualização dos resultados do Modelo KNN nos dados de teste

plt.rcParams['figure.figsize'] = (20,12)
plt.rcParams['font.size'] = 16
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['axes.labelsize'] = 20

tamanho = len(y_test)
plt.scatter([i for i in range(0,tamanho) if previsoes_knn[i] == y_test[i]], previsoes_knn[previsoes_knn==y_test], s=50, color='green', label='Previsões Corretas')
plt.scatter([i for i in range(0,tamanho) if previsoes_knn[i] != y_test[i]], previsoes_knn[previsoes_knn!=y_test], s=50, color='red', label='Previsões Incorretas')
plt.title('Resultado do Modelo KNN ({} primeiras instâncias)'.format(tamanho))
plt.xlabel('Instãncias')
plt.xticks(range(0, tamanho))
plt.xticks(rotation=90)
plt.yticks(range(0, 2))
plt.ylabel('Tipo: 2 - Benígno / 4 - Malígno')
plt.legend()
plt.show()
```

# Modelo Random Forest

```python
# Carregando o dataset

dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Dividindo o dataset em dados de treino e dados de teste

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 0)
```

```python
# Dimensionsamento de recursos

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Construção do Modelo Random Forest

from sklearn.ensemble import RandomForestClassifier
clf_random = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
```

```python
# Treinamento do Modelo Random Forest

clf_random.fit(X_train, y_train)
```

```python
# Previsão do Modelo Random Forest nos dados de teste

previsoes_random = clf_random.predict(X_test)
```

```python
# Matriz de Confusão do Modelo Random Forest

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, previsoes_random)
```

```python
# Acurácia do Modelo Random Forest

from sklearn.metrics import accuracy_score
accuracy_score(y_test, previsoes_random)
```

```python
# Visualização dos resultados do Modelo de Random Forest nos dados de teste

plt.rcParams['figure.figsize'] = (20,12)
plt.rcParams['font.size'] = 16
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['axes.labelsize'] = 20

tamanho = len(y_test)
plt.scatter([i for i in range(0,tamanho) if previsoes_random[i] == y_test[i]], previsoes_random[previsoes_random==y_test], s=50, color='green', label='Previsões Corretas')
plt.scatter([i for i in range(0,tamanho) if previsoes_random[i] != y_test[i]], previsoes_random[previsoes_random!=y_test], s=50, color='red', label='Previsões Incorretas')
plt.title('Resultado do Modelo Random Forest ({} primeiras instâncias)'.format(tamanho))
plt.xlabel('Instãncias')
plt.xticks(range(0, tamanho))
plt.xticks(rotation=90)
plt.yticks(range(0, 2))
plt.ylabel('Tipo: 2 - Benígno / 4 - Malígno')
plt.legend()
plt.show()
```

# Avaliação Geral

```python
# Conjunto de acurácias

acuracias = {}
acuracias['Regressão Logistica'] = accuracy_score(y_test, previsoes_logistica)
acuracias['KNN'] = accuracy_score(y_test, previsoes_knn)
acuracias['Kernel SVM'] = accuracy_score(y_test, previsoes_svm)
acuracias['Naive Bayes'] = accuracy_score(y_test, previsoes_svm)
acuracias['Árvore de Decisão'] = accuracy_score(y_test, previsoes_arvore)
acuracias['Random Forest'] = accuracy_score(y_test, previsoes_random)

df_acuracias = pd.DataFrame.from_dict(acuracias, orient='index').reset_index()
df_acuracias.columns=['Modelo','Acurácia']

grafico = df_acuracias.set_index('Modelo').sort_values(by='Acurácia',ascending=False)
```

```python
# Visualizando os resultados dos Modelos por ordem de acurácia

plt.rcParams['figure.figsize'] = (12,8)
plt.rcParams['font.size'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['axes.labelsize'] = 20

ax = grafico.Acurácia.plot(kind='bar', rot=30)
ax.set_yticks(np.arange(0, 1.05, step=0.05))
ax.set_ylim(0.85)
ax.set_title("Acurácia dos Modelos")
ax.set_ylabel("Acurácia nos dados de teste")

rects = ax.patches

labels = round(grafico.Acurácia,2)

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height, label, ha='center', va='bottom')    
    
plt.show()
```

```python
# Visualização dos resultados dos Modelos nos dados de teste

plt.rcParams['figure.figsize'] = (20,12)
plt.rcParams['font.size'] = 16
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['axes.labelsize'] = 20

tamanho = len(y_test)
s=100

plt.scatter([i for i in range(0,tamanho) if previsoes_logistica[i] != y_test[i]], previsoes_logistica[previsoes_logistica!=y_test]+0.05, s=s, marker='*', color='green', label='Regressão Logística')
plt.scatter([i for i in range(0,tamanho) if previsoes_knn[i] != y_test[i]], previsoes_knn[previsoes_knn!=y_test]+0.1, s=s, marker='D', color='pink', label='KNN')
plt.scatter([i for i in range(0,tamanho) if previsoes_svm[i] != y_test[i]], previsoes_svm[previsoes_svm!=y_test]+0.15, s=s, marker='s', color='purple', label='SVM')
plt.scatter([i for i in range(0,tamanho) if previsoes_naive[i] != y_test[i]], previsoes_naive[previsoes_naive!=y_test]-0.15, s=s, marker='p', color='brown', label='Naive Bayes')
plt.scatter([i for i in range(0,tamanho) if previsoes_arvore[i] != y_test[i]], previsoes_arvore[previsoes_arvore!=y_test]-0.1, s=s, marker='^', color='red', label='Árvore de Decisão')
plt.scatter([i for i in range(0,tamanho) if previsoes_random[i] != y_test[i]], previsoes_random[previsoes_random!=y_test]-0.15, s=s, marker='o', color='blue', label='Random Forest')

plt.title('Previsões incorretas de cada Modelo({} primeiras instâncias)'.format(tamanho))
plt.xlabel('Instãncias')
plt.xticks(np.arange(0, tamanho, step=2), rotation=45)
plt.yticks(np.arange(2, 5, step=2))
plt.ylabel('Previsão do Modelo: 2 - Benígno / 4 - Malígno')
plt.legend(loc='center')
plt.show()
```

```python
# Dados de previsões corretas e incorretas

df_acuracias['Corretas'] = 0
df_acuracias['Incorretas'] = 0

df_acuracias['Corretas'] = [int(i) for i in len(y_test)*df_acuracias.Acurácia]
df_acuracias['Incorretas'] = [int(i) for i in len(y_test)*(1-df_acuracias.Acurácia)]
```

```python
# Visualizando a quantidade de previsões corretas e incorretas por modelo

plt.rcParams['figure.figsize'] = (12,8)
plt.rcParams['font.size'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['axes.labelsize'] = 20

ax = df_acuracias.set_index('Modelo').sort_values(by='Corretas', ascending=False).iloc[:,1:].plot(kind='bar', rot=30)
ax.set_yticks(np.arange(0, len(y_test), step=10))
ax.set_title("Previsões Corretas x Incorretas")
ax.set_ylabel("Quantidade de instâncias")
ax.legend(loc='center')

rects = ax.patches

listas = [[i for i in df_acuracias.sort_values(by='Corretas', ascending=False).Corretas], [j for j in df_acuracias.sort_values(by='Corretas', ascending=False).Incorretas]]

labels = [val for sublist in listas for val in sublist]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height, label, ha='center', va='bottom')    
    
plt.show()
```
