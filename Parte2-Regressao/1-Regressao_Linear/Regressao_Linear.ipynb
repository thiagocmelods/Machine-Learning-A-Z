```python
# Importação de Bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando o dataset

dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Dividindo o dataset em dados de treino e dados de teste

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python
# Construção do Modelo

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
# Treinamento

regressor.fit(X_train, y_train)
```

```python
# Previsão nos dados de teste

previsoes = regressor.predict(X_test)
```

```python
# Visualização dos resultados (em dados de treino)

plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("Salário x Anos de Experiência (Dados de Treino)")
plt.xlabel("Anos de Experiência")
plt.ylabel("Salário")
plt.show()
```

```python
# Visualização dos resultados (em dados de teste)

plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("Salário x Anos de Experiência (Dados de Teste)")
plt.xlabel("Anos de Experiência")
plt.ylabel("Salário")
plt.show()
```

```python
# Inserindo um valor de anos de experiência para prever o salário

ponto = float(input("Digite o número de anos de experiência para prever o salário: "))
```

```python
# Visualização dos resultados

plt.scatter(X_train, y_train, color = 'red', label='Treino')
plt.scatter(X_test, y_test, color = 'green', label='Teste')
plt.scatter([[ponto]], regressor.predict([[ponto]]), color = 'black', label="Ponto específico: {} anos" .format(ponto), s=60)
plt.plot(X_train, regressor.predict(X_train), color = 'blue', label="Linha de regressão")
plt.title("Salário x Anos de Experiência")
plt.xlabel("Anos de Experiência")
plt.ylabel("Salário")
plt.legend()
plt.show()
```
