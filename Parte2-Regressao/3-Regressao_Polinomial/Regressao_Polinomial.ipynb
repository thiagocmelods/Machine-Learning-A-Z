```python
# Importação de Bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando o dataset

dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Construção do Modelo de Regressão Linear

from sklearn.linear_model import LinearRegression
regressor_linear = LinearRegression()
```

```python
# Treinamento do Modelo de Regressão linear

regressor_linear.fit(X, y)
```

```python
# Construção do Modelo de Regressão Polinomial

from sklearn.preprocessing import PolynomialFeatures
poli = PolynomialFeatures(degree=2)
```

```python
# Treinamento do Modelo de Regressão Polinomial

X_poli = poli.fit_transform(X)
regressor_poli = LinearRegression()
regressor_poli.fit(X_poli, y)
```

```python
# Visualização dos resultados da Regressão linear

plt.scatter(X, y, color = 'red')
plt.plot(X, regressor_linear.predict(X), color = 'blue')
plt.title("Verdade ou Blefe (Regressão Linear)")
plt.xlabel("Posição")
plt.ylabel("Salário")
plt.show()
```

```python
# Visualização dos resultados da Regressão Polinomial

plt.scatter(X, y, color = 'red')
plt.plot(X, regressor_poli.predict(X_poli), color = 'blue')
plt.title("Verdade ou Blefe (Regressão Polinomial)")
plt.xlabel("Posição")
plt.ylabel("Salário")
plt.show()
```

```python
# Inserindo uma posição para prever o salário

ponto = float(input("Digite o nível da posição para prever o salário: "))
```

```python
# Visualização dos resultados da Regressão Polinomial

X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red', label='Real')
plt.scatter([[ponto]], regressor_poli.predict(poli.fit_transform([[ponto]])), label='Ponto Específico: {}'.format(ponto), s=100, color='black')
plt.plot(X_grid, regressor_poli.predict(poli.fit_transform(X_grid)), color = 'blue', label='Previsto')
plt.title("Verdade ou Blefe (Regressão Polinomial)")
plt.xlabel("Posição")
plt.ylabel("Salário")
plt.legend()
plt.show()
```
